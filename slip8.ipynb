```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv('C:\\Users\\Admin\\Desktop\\Groceries data.csv')
df 
```

```python
records=[]
for i in range(0,38765):
    records.append([str(df.values[i,j]) for j in range(0,7)])
```

```python
records
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
te_array=te.fit(records).transform(records)
```

```python
df=pd.DataFrame(te_array,columns=te.columns_)
```

```python
df
```

```python
freq_items=apriori(df,min_support=0.0045,use_colnames=True)
```

```python
print(freq_items)
```

```python
rules=association_rules(freq_items,metric='support',min_threshold=0.0045)
rules=rules.sort_values(['support','confidence'],ascending=[False,False])
```
